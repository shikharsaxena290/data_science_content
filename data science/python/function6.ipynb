```python
def table(number,n):
  no=1
  result=0
  while no<=n:
    result=number*no
    print(number,"*",no,"=",result)
    no=no+1
table(5,20)
```

```python
table(19,10)
```

```python
l=[1,2,3,4,5,"shikhar","saxena","pwskills",22.34,43.555,[5,6,7,8]]
def listt(l):
  l1=[]
  l2=[]
  for i in l:
    if type(i)==list or type(i)==int or type(i)==float:
      l1.append(i)
    else:
      l2.append(i)
  return l1,l2
listt(l)
```

```python
def test11(*args):  #by using *, n no's of attributes u can pass in function called steric and always give output in tuples
  return args
test11(1,2,3)
```

```python
def test12(*args,a):
  return args,a
test12(1,2,3,4) #creats confusion which value is for a to avoid this use syntax given in next code
```

```python
test12(1,2,3,a=4)
```

```python
def test13(**kwargs):        #in case of **(double steric) kwargs output is in form of dictionary
  return kwargs
test13()
```

```python
test13(a=[1,2,3,4],b=[3,4,5,6],c=[6,7,8,9])
```
