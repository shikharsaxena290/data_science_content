## **Generator function**

```python
def test_fib(n):
  a,b=0,1
  for i in range(n):
    yield a
    a,b=b,a+b
```

```python
for i in test_fib(10):
  print(i)
```


In Python, a generator function is a special type of function that allows you to iterate over a potentially large sequence of items without creating the entire sequence in memory at once. It generates values one at a time, on-the-fly, as you iterate over it. This can be more memory-efficient compared to creating a list or other iterable with all the values upfront.

The key difference between a regular function and a generator function is that a generator function uses the yield keyword instead of return. When the generator function encounters a yield statement, it temporarily suspends its execution and yields a value to the caller. The next time the generator is called, it resumes from where it was suspended, keeping track of its state.

```python
def test_fib1():
  a,b=0,1
  while True:
    yield a
    a,b=b,a+b
```

```python
fib = test_fib1()
for i in range(10):
  print(next(fib)) #next() used to return the next item from the iterator
```

```python
def count_test(n):
  count = 1
  while count<n:
    yield count
    count = count + 1
```

```python
c= count_test(5)
for i in c:
  print(i)
```

## **Lambda Function**

```python
n = 3
p = 2
def test(n,p):
  return n**p
test(3,2)
```

```python
#writing in form of lambda function also known as one liner function or anonymous function
a = lambda n,p : n**p
a(3,2)
```

```python
add = lambda x,y:x+y
add(4,5)
```

```python
ferhn = lambda c :c * (9/5) + 32
ferhn(100)
```

```python
finding_max = lambda x,y : x if x>y else y
finding_max(24,34)
```

```python
s = "pwskills"
find_len = lambda s : len(s)
find_len(s)
```
