```python
result=0
n= int(input("enter no upto which u want sum "))
while n>0:
  result=result+n
  n=n-1
result
```

```python
n=int(input("enter no "))
factorial=1
while n>1:
  factorial=factorial*n
  n=n-1
factorial
```

```python
n=int(input("enter no upto which u want fibbonaci series "))
a=0
b=1
counter= 0
while counter<n:
  print(a)
  c=a+b
  a=b
  b=c
  counter=counter+1
```

```python
str=input("enter your string")
reverse=""
for i in str:
  reverse = i+reverse
print(reverse)
```

```python
str=input("enter your string ") #ram
reverse="" #""
length = len(str) #3
while length>0: #3>0 true
  reverse=reverse + str[length -1 ] #"" + str[3-1 = 2] ie str[2] = m therefore reverse="m"
  length = length-1 #3-1 = 2
print(reverse)
```

```python
number=int(input("enter number ")) #5
table = 1
result=0
while table<=10: #1<=10 true , 2<=10,3<=10
  result=number*table  #result = 5*1 = 5 , 5*2=10,5*3=15
  print(number,"*",table,"=",result) #5,10,15
  table=table+1 #updated value = 2,3,4
```
