```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,200)
x
```

```python
y=np.sin(x)
y
```

```python
plt.plot(x,y) #sin graph
```

```python
#adding graph x,y label and graph title
plt.xlabel("x data")
plt.ylabel("y data")
plt.title("sin graph")
plt.plot(x,y)
```

```python
plt.xlabel("x data")
plt.ylabel("y data")
plt.title("sin graph")
plt.plot(x,y ,'--r')
```

```python
plt.xlabel("x data")
plt.ylabel("y data")
plt.title("sin graph")
plt.plot(x,y ,'--g')
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,y ,'--g')
plt.xlabel("x data")
plt.ylabel("y data")
plt.title("sin graph")
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,y ,'--g')
plt.xlabel("x data")
plt.ylabel("y data")
plt.title("sin graph")
plt.grid()
```

```python
x=np.random.rand(50)
x
```

```python
y=np.random.rand(50)
y
```

```python
plt.scatter(x,y)
```

```python
#adding colors
colours=np.random.rand(50)
plt.scatter(x,y ,c=colours)
```

```python
#controling size
colours=np.random.rand(50)
sizes=np.random.rand(50)
plt.scatter(x,y ,c=colours , s=sizes)
```

```python
#increasing size
colours=np.random.rand(50)
sizes=50*np.random.rand(50)
plt.scatter(x,y ,c=colours , s=sizes)
```

```python
#aplha represent intensity 
colours=np.random.rand(50)
sizes=50*np.random.rand(50)
plt.scatter(x,y ,c=colours , s=sizes ,alpha=.5)
```

```python
#adding label and title
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("scatter graph")
#adding colour size and alpha
colours=np.random.rand(50)
sizes=50*np.random.rand(50)
plt.scatter(x,y ,c=colours , s=sizes ,alpha=.5)
```

```python
#adding label and title
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("scatter graph")
#adding colour size and alpha and marker
colours=np.random.rand(50)
sizes=50*np.random.rand(50)
plt.scatter(x,y ,c= "#FFBB33", s=sizes ,alpha=.5 , marker='x') #giving color code
```

```python
#adding label and title
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("scatter graph")
#adding colour size and alpha and marker
colours=np.random.rand(50)
sizes=50*np.random.rand(50)
plt.scatter(x,y ,c= "#FFBB33", s=sizes ,alpha=.5 , marker='v') #giving color code
```

```python
x=[1,2,3,4,5]
y=[1,4,6,8,10]
colour=['red','blue','green','black','yellow']
plt.scatter(x,y , c=colour)
```

```python
#categorical vs numerical data
x=['a','b','c','d','e']
y=np.random.rand(5)
y
```

```python
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar graph")
plt.bar(x,y)
```

```python
#making graph horizontal
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar graph")
plt.barh(x,y)
```

```python
#repeating data
data=[1,2,3,4,2,4,3,1,6,7,8,1,2,3]
plt.hist(data)
```

```python
#3d plot
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z)
```
