SMOTE(Synthetic Minority oversampling technique)

SMOTE is a technique uesed in machine learning to address imbalanced datasets where the minorities class has significantly fewer instances than the majority class . SMOTE involves generating synthetic instances of the minority class by interpolating between existing instances

```python
from sklearn.datasets import make_classification 
```

```python
#x is independent feature
#y is dependent feature
x,y=make_classification(n_samples=1000 , n_features=2,n_redundant=0,n_clusters_per_class=1, weights=[0.90])
```

```python
import pandas as pd
```

```python
df1=pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
```

```python
final_df
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```

DATA Interpolation

data interpolation is the process of estimating unknown values within a dataset based on the known values. in python there are various libraries available that can be used for data interpolation, such as numpy ,scipy and pandas. here is an example of how to perform data interpolation using the numpy library

1 linear interpolation

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
```

```python
#interpolate the data using linear interpolation
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

```python
#Interpolating the data using linear interpolation
x_new = np.linspace(1,5,15)##creating new x values
y_interp = np.interp(x_new , x,y)#interpolate y values
print(y_interp)
```

```python
plt.scatter(x_new , y_interp)
```

2 cubic interpolation with scipy

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125])
```

```python
from scipy.interpolate import interp1d
```

```python
#create cubic interpolation
f=interp1d(x,y,kind='cubic')
```

```python
x_new = np.linspace(1,5,10)
y_interp =f(x_new)
print(y_interp)
```

```python
plt.scatter(x,y)#graph before interpolation
```

```python
plt.scatter(x_new,y_interp)#graph after interpolation
```

3 polynomial interpolation

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,4,9,16,25])
```

```python
#interpolate the data using polynomial interpolate
p=np.polyfit(x,y,2)#fit a 2nd degree polynomial to the data
```

```python
x_new = np.linspace(1,5,10)
y_interp = np.polyval(p,x_new)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```
